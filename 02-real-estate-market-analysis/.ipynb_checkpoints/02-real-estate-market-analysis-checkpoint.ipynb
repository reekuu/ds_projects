# Исследование объявлений о продаже квартир

**Описание проекта**

В нашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Наша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Описание данных**

Вписаны пользователем:

* __balcony__ — число балконов
* __ceiling_height__ — высота потолков (м)
* __days_exposition__ — сколько дней было размещено объявление (от публикации до снятия)
* __first_day_exposition__ — дата публикации
* __floor__ — этаж
* __floors_total__ — всего этажей в доме
* __is_apartment__ — апартаменты (булев тип)
* __kitchen_area__ — площадь кухни в квадратных метрах (м²)
* __last_price__ — цена на момент снятия с публикации
* __living_area__ — жилая площадь в квадратных метрах(м²)
* __locality_name__ — название населённого пункта
* __open_plan__ — свободная планировка (булев тип)
* __rooms__ — число комнат
* __studio__ — квартира-студия (булев тип)
* __total_area__ — площадь квартиры в квадратных метрах (м²)
* __total_images__ — число фотографий квартиры в объявлении

_Пояснение: апартаменты — это нежилые помещения, не относящиеся к жилому фонду, но имеющие необходимые условия для проживания._

Получены автоматически на основе картографических данных:

* __airports_nearest__ — расстояние до ближайшего аэропорта в метрах (м)
* __cityCenters_nearest__ — расстояние до центра города (м)
* __parks_around3000__ — число парков в радиусе 3 км
* __parks_nearest__ — расстояние до ближайшего парка (м)
* __ponds_around3000__ — число водоёмов в радиусе 3 км
* __ponds_nearest__ — расстояние до ближайшего водоёма (м)

**Выделение важных данных**

Значения, которые предстоит высчитать, и необходимые для этого столбцы:

* цена квадратного метра (`last_price`, `total_area`);
* день недели, месяц и год публикации объявления (`first_day_exposition`);
* этаж квартиры; варианты — первый, последний, другой (`floor`, `floors_total`);
* соотношение жилой и общей площади, а также отношение площади кухни к общей (`total_area`, `living_area`, `kitchen_area`).

Важные признаки:

* число комнат (`rooms`);
* высота потолков (`ceiling_height`);
* время продажи квартиры (`days_exposition`);
* удалённость от центра (`cityCenters_nearest`);
* название населённого пункта (`locality_name`).

### Поиск аномалий и ошибок, изучение пропусков

```python
# инициализируем функцию na_percent() для вывода красивых сообщений о количестве пропусков
def na_percent(column):

    na = df[df[column].isna()].shape[0]
    percent = na / df.shape[0]
    print(f"В {column} пропущено {na} значений, это {percent:.2%} от общего объема данных.")
```

**Цена `last_price`**

```python
# выведем таблицу с базовыми статистическими метриками в last_price
display(df['last_price'].describe())
```

Минимальное значение сильно отличается от первого квартиля. Посмотрим на него отдельно.

```python
# выведем строку с этим значением целиком
display(df.query('last_price == last_price.min()'))
```

Похоже, что цена указана в млн. рублей, в то время как остальные цены просто в рублях. Исправим это на следующем шаге.

**Общая площадь `total_area`, жилая `living_area` и кухонная `kitchen_area`**

```python
# выведем таблицу с базовыми статистическими метриками в total_area, living_area, kitchen_area
display(df[['total_area', 'living_area', 'kitchen_area']].describe())
```

Максимальные значения сильно превышают третий квартиль во всех трех столбцах, хотя остальные значения выглядят преимущественно нормально. Проверим есть ли случаи, когда жилая площадь или площадь кухни превышает общую.

```python
display(df.query('living_area > total_area or kitchen_area > total_area'))
```

Таких случаев нет.

```python
# подсчитаем число пропусков в столбце living_area
na_percent('living_area')

# подсчитаем число пропусков в столбце kitchen_area
na_percent('kitchen_area')
```

```python
# подсчитаем число пропусков в столбце days_exposition
na_percent('days_exposition')
```

Из пояснения к данным следует, что признак сообщает сколько дней было размещено объявление (от публикации до снятия). Есть основание полагать, что пропуски указывают на актуальные объявления, то есть еще не снятые.

Основная задача проекта – научиться определять рыночную стоимость объектов недвижимости. Мы будем действовать на основе допущения о том, что снятые с публикации объявления соответствуют проданным квартирам. А для установления розничной цены лучше оперировать данными именно о таких квартирах.

По СНиП высота потолков в жилых помещениях должна быть не меньше 2.5 метров. Гистограмма демонстрирует, что в жизни бывает иначе, но значения ниже 2.2 всё равно похожи на ошибку. Также маловероятно встретить 8 метровые потолки где бы то ни было. Последняя строка таблицы со значением `ceiling_height == 100` и вовсе целиком похожа на фейк.

```python
# инициализируем функцию для помывки потолков
def wash(x):

    # уменьшим значения в интервале от 22 до 32 на порядок
    if 22 <= x <= 32:
        return x/10

    # заменим маловероятные значения на None
    elif x < 2.2 or 8 <= x < 22:
        return None

    # остальные значения оставим как есть
    else:
        return x
```

```python
# удалим строку похожую на фейк
df = df[df['ceiling_height'] != 100.0]

# применим функцию к ceiling_height
df['ceiling_height'] = df['ceiling_height'].apply(lambda x: wash(x))

# насладимся чистыми потолками
display(df['ceiling_height'].round().value_counts())
```

**Высота потолка `ceiling_height`**

Выполним предварительную подготовку.

```python
# инициализируем функцию для расчёта округлённого расстояния до центра Санкт-Петербурга
def spb_km_to_center(row):

    # зададим короткие имена для обращения к locality_name и city_centers_nearest
    loc = row['locality_name']
    distance = row['city_centers_nearest']

    # для Санкт-Петербурга заполним признак округленным до км расстоянием
    if loc == "Санкт-Петербург" and distance < 27000:
        return int(round(distance/1000))

    # для остальных локаций заполним признак индикаторным значением
    return 999
```

Добавим в таблицу новый столбец и заполним пропуски.

```python
# заполним столбец spb_km_to_center применив одноименную функцию к таблице df методом apply()
df['spb_km_to_center'] = df.apply(spb_km_to_center, axis=1)

# скопируем оригинальный столбец ceiling_height для сравнения в конце
original_ceiling_height = df['ceiling_height'].copy()

# заполним пропуски в ceiling_height с помощью transform() на основе соответствующего медианного значения
df['ceiling_height'].fillna(df
                            .groupby(['locality_name', 'spb_km_to_center'])['ceiling_height']
                            .transform('median'), inplace=True)
```

**Добавление категории этажа — первый, последний, другой**

```python
# инициализируем функцию для удобной категоризации методом apply()
def floor_cat(row):
    
    # присвоим нужным ячейкам короткие имена
    floor = row['floor']
    total = row['floors_total']

    # сформируем вывод на основе условия
    if floor == 1 or total == 1:
        return 'первый'
    
    if floor == total:
        return 'последний'

    return 'другой'

# протестируем функцию
print(floor_cat({'floor': 1, 'floors_total': 10}))
print(floor_cat({'floor': 5, 'floors_total': 10}))
print(floor_cat({'floor': 10, 'floors_total': 10}))
```

```python
# создадим новый столбец в df и заполним его
df['floor_cat'] = df.apply(lambda row: floor_cat(row), axis=1)

# выведем 10 случайных строк из датасета для проверки
display(df[['floor', 'floors_total', 'floor_cat']].sample(10))
```

**Расчет и добавление соотношения жилой и общей площади, а также отношение площади кухни к общей**

```python
# добавим столбец с соотношением жилой площади к общей, округлим до 2 знаков после запятой
df['living_area_ratio'] = round(df['living_area'] / df['total_area'], 2)

# добавим столбец с соотношением площади кухни к общей, округлим до 2 знаков после запятой
df['kitchen_area_ratio'] = round(df['kitchen_area'] / df['total_area'], 2)

# проверим результат
display(df[['total_area', 'living_area', 'living_area_ratio',
            'kitchen_area', 'kitchen_area_ratio']].sample(5))
```

Заведем функцию для автоматизации процесса построения гистограмм.

Для левой границы графика используем минимальное значение признака, так как по описанию наши данные имеют положительную ассиметрию и не имеют отрицательных значений. Для правой границы графика используем простейшую формулу, основанную на свойствах нормального распределения: `Upper Bound: (Q3 + 1.5 * IQR)`

Будем считать выбросом всё, что превышает значение суммы 3-го квартиля и полутора межквартильных расстояния. Другими словами, воспользуемся *правилом трёх сигм*.

**Правило трех сигм (3-sigma rule)** – это правило, утверждающее, что вероятность того, что случайная величина отклонится от своего математического ожидания более чем на три среднеквадратических отклонения, практически равна нулю. В пределах двух среднеквадратических отклонений — уже 95,44%, а в пределах трех — 99,72%.

```python
# инициализируем функцию для быстрого вывода гистограммы
def show_hist(df, column, title, xlabel):

    # подготовим значения
    src = df[column]
    minimum = src.min()
    q1 = round(src.quantile(0.25), 2)
    q2 = round(src.quantile(0.50), 2)
    q3 = round(src.quantile(0.75), 2)
    iqr = q3-q1
        
    # зададим левую границу графика по минимуму
    r_min = round(minimum)

    # зададим правую правую границу графика на основе формулы, округлим по верхней границе
    r_max = ceil(q3 + 1.5*iqr)

    # зададим количество корзин равное размаху графика
    bins = r_max - r_min
    
    # внесем корректировку для малого числа корзин (подставим костыли)
    if bins < 5:
        bins *= 10

    # построим гистограмму
    df.plot(kind='hist', y=column, bins=bins, range=(r_min, r_max), alpha=0.8)

    # зададим заголовок и названия осей
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Частота признака')

    # выведем показатели левого и правого края графика
    tab = ' '*13
    print()
    print()
    print(f"{tab}Min: {r_min}{tab}25%: {q1}{tab}50%: {q2}{tab}75%: {q3}{tab}Max: {r_max}")
    
    # выведем график
    return plt.show()
```

```python
# заведем список с пресетом для функции show_hist()
preset = [
        {'column': 'total_area',
        'title': 'Распределение общей площади квартир',
        'xlabel': 'Площадь в квадратных метрах'},

        {'column': 'last_price',
        'title': 'Распределение стоимости квартир (на момент снятия с публикации)',
        'xlabel': 'Цена в млн. руб.'},

        {'column': 'rooms',
        'title': 'Распределение количества комнат в продаваемых квартирах',
        'xlabel': 'Количество комнат'},

        {'column': 'ceiling_height',
        'title': 'Распределение высоты потолков в продаваемых квартирах',
        'xlabel': 'Высота в метрах'}]
```

```python
# выведем гистограммы для столбцов из списка preset
for item in preset:
    show_hist(df.query('days_exposition > -1'), **item)
```

```python
# создадим отдельный датафрейм
df_clean_saintp = df_clean.query('locality_name == "Санкт-Петербург" and days_exposition < 999').copy()

# подготовим сводную таблицу и сформируем график
(df_clean_saintp
 .pivot_table(index='spb_km_to_center', values='last_price')
 .plot(kind='bar'))

# добавим заголовок и названия осей
plt.title('Зависимость средней стоимости жилья от расстояния до центра города')
plt.xlabel('Растояние до центра в километрах')
plt.ylabel('Цена в млн. руб.')

# выведем график
plt.show()
```

Их границы можно грубо очертить кругом с центром у Эрмитажа и радисом около 2-3 км. По окраине этого радиуса идёт активное строительство, преимущественно в бывших промышленных зонах, их еще называют "промышленным поясом". Таким образом, провал на графике можно попробовать объяснить тем, что еще не все промышленные зоны расчищены и застроены.

### Анализ центральной зоны города Санкт-Петербурга и сравнение с общими параметрами

Заведем столбец `spb_city_center` с флагом **True** для центральной зоны и флагом **False** для всего остального.

```python
# заполним столбец с помощью анонимной лямбда-функции
df_clean_saintp['spb_city_center'] = df_clean_saintp.apply(
    lambda row: True if row.spb_km_to_center <= 8 else False, axis=1)
```

**Изучение зависимости цены `last_price` от площади `total_area`**

```python
# построим график
plot = sns.jointplot(x="total_area", y="last_price",
                     data=df_clean_saintp.query('spb_city_center == True'),
                     kind='hist')

# добавим заголовок и подгоним размер графика
plot.fig.suptitle('Зависимость цены и общей площади')
plot.fig.subplots_adjust(top=0.95)
plot.fig.tight_layout()

# выведем график
plt.show()

# рассчитаем коэф. корреляции Пирсона
display(round(df_clean_saintp.query('spb_city_center == True')
              [['last_price', 'total_area']].corr(), 2))
```

* **Коэффициент корреляции Пирсона равен 0.71** – высокая положительная связь.

Распределение имеет диагональную направленность от низких значений к высоким.

* Фактически график напоминает общую картину, с той разницей, что здесь самые популярные квартиры имеют площадь 40-60 кв.м. цену 5-7 млн. руб.

**Изучение зависимости цены `last_price` от числа комнат `rooms`**

```python
# построим график
sns.lineplot(x="rooms", y="last_price",
             data=df_clean_saintp.query('spb_city_center == True'),
             palette='Set2')

# зададим заголовок и названия осей
plt.title('Средняя стоимость квартир для различного числа комнат')
plt.xlabel('Количество комнат')
plt.ylabel('Цена в млн. руб.')

# выведем график
plt.show()

# рассчитаем коэф. корреляции Пирсона для тех же данных
display(round(df_clean_saintp.query('spb_city_center == True')
              [['last_price', 'rooms']].corr(), 2))
```

* **Коэффициент корреляции Пирсона равен 0.44** – слабая положительная связь (чем больше комнат, тем выше cредняя цена). 

Данных по 5-комн. квартирам мало, поэтому доверительный интервал на графике шире (обозначен красной тенью за кривой среднего). Распределение также имеет диагональную направленность от низких значений в высоким. 

* 1 комн. квартира имеет среднюю цену 5.6 млн. рублей;
* 2 комн. квартира имеет среднюю цену 7 млн. рублей;
* 3 комн. квартира имеет среднюю цену 8 млн. рублей;
* 4 комн. квартира имеет среднюю цену 8,8 млн. рублей;
* для 5 комн. и 6 комн. квартир слишком мало данных;

Средняя цена для центра Санкт-Петербурга почти на 2 млн. рублей выше, чем для региона в целом.

**Изучение зависимости цены квадратного метра `sqm_price` от удалённости от цента `city_centers_nearest`**

```python
# построим график
plot = sns.jointplot(x="city_centers_nearest", y="sqm_price",
                     data=df_clean_saintp.query('spb_city_center == True'),
                     kind='hist')

# добавим заголовок и подгоним размер графика
plot.fig.suptitle('Зависимость цены кв.м. от удаленности до центра')
plot.fig.subplots_adjust(top=0.95)
plot.fig.tight_layout()

# выведем график
plt.show()

# рассчитаем коэф. корреляции Пирсона для тех же данных
display(round(df_clean_saintp.query('spb_city_center == True')
              [['sqm_price', 'city_centers_nearest']].corr(), 2))
```

Однако основная масса рассеяна между приблизительн 70 и 200 тысячами за квадратный метр. График показывает, что внутри центральной зоны удаленность от центра не имеет влияния на цену квадратного метра.

**Изучение зависимости цены `last_price` от категории этажа `floor_cat`**

```python
# построим график boxplot
sns.boxplot(y="floor_cat", x="last_price",
            data=df_clean_saintp.query('spb_city_center == True'),
            whis=inf, palette='Set2')

# поверх него построим stripplot
sns.stripplot(y="floor_cat", x="last_price",
              data=df_clean_saintp.query('spb_city_center == True'),
              color=".5", alpha=0.03, jitter=0.4, palette='Set2')

# зададим заголовок и названия осей
plt.title('Распределение средней стоимости квартир в зависимости от категории этажа')
plt.xlabel('Цена в млн. руб.')
plt.ylabel('Категория этажа')

# выведем графики
plt.show()

# выведем сводную таблицу для информативности
display(round(df_clean_saintp.query('spb_city_center == True')
              .pivot_table(index='floor_cat', values='last_price', aggfunc='median')
              .sort_values(by='last_price'), 2))
```

В то же время разница между последним и другим этажом почти исчезла и составляет лишь 200 тысяч. Жители центра явно больше любят последние этажи.
